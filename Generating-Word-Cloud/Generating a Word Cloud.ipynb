## Create a wordcloud using the `wordcloud` library and the "Women's E-Commerce Clothing Reviews" dataset from Kaggle.

Link to the dataset - https://www.kaggle.com/datasets/nicapotato/womens-ecommerce-clothing-reviews

```python
# Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv("Womens Clothing E-Commerce Reviews.csv")
df.head()
```

```python
df.info()
```

```python
# Checking null values
df.isna().sum()
```

```python
df.dropna(subset=['Review Text'], inplace=True)
df.isna().sum()
```

```python
stopwords = set(STOPWORDS)

#Creating the text variable
text1 = " ".join(txt for txt in df['Review Text'])

# Generate word cloud 1
wordcloud = WordCloud(
    width=3000,
    height=2000,
    random_state=1,
    background_color="black",
    colormap="Pastel1",
    collocations=False,
    stopwords=stopwords,
    ).generate(text1)

# Plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
